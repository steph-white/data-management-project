````python
#load csv file
df = pd.read_csv('database_24_25.csv')

df.head()
````

````python
df.info()
````

````python
#drop rows that do not include ORL or OKC
filtered_df = df[df['Tm'].isin(['ORL', 'OKC'])]
filtered_df.shape
````

````python
messy_df = filtered_df.copy()
````

````python
dbl_row = messy_df[messy_df['Player'] == 'Franz Wagner']  #making every row w/ franz wagners name to be duplicated
messy_df = pd.concat([messy_df, dbl_row], ignore_index = True)  #duplicating and relabeling index
````

````python
messy_df.loc[0:2, 'AST'] = np.nan #sets first 3 rows of assists to missing
messy_df.loc[5:7, '3P%'] = np.nan #sets rows 5-7 of 3pt% to missing
````

````python
messy_df.to_csv("nba_24_25_messy.csv", index=False)
````
